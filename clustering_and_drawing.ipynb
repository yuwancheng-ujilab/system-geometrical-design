```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import random
import os
from mpl_toolkits.mplot3d import Axes3D
#import kmeans#
from sklearn.cluster import KMeans
%matplotlib notebook

#set the path for this code
path=r'xxx'
os.chdir(path)

#load pareto data
paretofront = np.loadtxt(r'.\xxx.csv',dtype="float",delimiter=",")
#color
colors=['y', 'm', 'k', 'b', 'r', 'c', 'g']
kmeans=[]

kmeans_color=KMeans(n_clusters=6).fit_predict(paretofront)
for i in kmeans_color:
    kmeanssingle=colors[i]
    kmeans.append(kmeanssingle)

coo_X = paretofront[:,7]    
coo_Y = paretofront[:,9]    
coo_Z = paretofront[:,8]    



data_all=np.column_stack((paretofront,kmeans_color))

fig = plt.figure()

ax = fig.add_subplot(221, projection='3d')
ax.scatter(coo_X, coo_Y,coo_Z,marker='o',c=kmeans,s=0.3)
ax.view_init(elev=20,azim=130)
ax.set_xlabel('f1(x)')
ax.set_ylabel('f2(x)')
ax.set_zlabel('f3(x)')


plt.subplot(222)
plt.scatter(coo_X, coo_Y,marker='o',c=kmeans,s=0.3)
plt.title("f1-f2")
plt.xlabel('f1(x)')
plt.ylabel('f2(x)')



plt.subplot(223)
plt.scatter(coo_X, coo_Z,marker='o',c=kmeans,s=0.3)
plt.title("f1-f3")
plt.xlabel('f1(x)')
plt.ylabel('f3(x)')


plt.subplot(224)
plt.scatter(coo_Y, coo_Z,marker='o',c=kmeans,s=0.3)
plt.title("f2-f3")
plt.xlabel('f2(x)')
plt.ylabel('f3(x)')



fig.tight_layout(pad=0.4, w_pad=4.0, h_pad=1.0)

patches = [mpatches.Patch(color=colors[i], label="{l}".format(l=i+1) ) for i in range(6) ]
plt.legend(handles=patches, bbox_to_anchor=(1.05, 1.5), loc=2, borderaxespad=0. )
fig.subplots_adjust(right=0.9)

plt.show()

plt.savefig("contourf_1001.png", dpi=300,bbox_inches = 'tight')





```
